```python
import pandas as pd
xl = pd.ExcelFile("National Election India 1977 - 2014.xlsx")
xl.sheet_names
```

```python
df = xl.parse("Sheet2")
df
```

```python
xl.sheet_names
```

```python
df = df.replace(['Chhattisgarh','Goa Daman & Diu', 'National Capital Territory Of Delhi', 'Nct Of Delhi','Orissa','Pondicherry','Uttaranchal'], ['Chattisgarh', 'Goa, Daman & Diu', 'Delhi', 'Delhi','Odisha','Puducherry','Uttarakhand'])
south_india=('Andhra Pradesh','Lakshadweep','Andaman & Nicobar Islands','Kerala','Karnataka', 'Pondicherry','Puducherry','Tamil Nadu')
north_india=('Rajasthan','Uttar Pradesh','Haryana','Punjab','Jammu & Kashmir','Delhi','Uttarakhand','Himachal Pradesh','Jharkhand', 'Dadra & Nagar Haveli','Chandigarh')
central_india=('Madhya Pradesh','Chattisgarh')
west_india=('Goa','Daman & Diu','Goa, Daman & Diu', 'Gujarat','Maharashtra')
east_india=('Bihar','Jharkhand','Odisha','Orissa','West Bengal','Arunachal Pradesh','Assam','Nagaland','Manipur', 'Mizoram','Meghalaya','Sikkim', 'Tripura')

df['region']='default'
df['region'].loc[df['st_name'].isin(south_india)]= 'south_india'
df['region'].loc[df['st_name'].isin(north_india)]= 'north_india'
df['region'].loc[df['st_name'].isin(central_india)]= 'central_india'
df['region'].loc[df['st_name'].isin(west_india)]= 'west_india'
df['region'].loc[df['st_name'].isin(east_india)]= 'east_india'

df
```

```python
# df.loc[df['column_name'] == some_value]
fdf = df.loc[df['cand_sex'] == 'F']
fdf
```

```python
# df.loc[df['year'] == 1977 and df['cand_sex'] == 'F']
sum(df.query("year == 1977 & cand_sex == 'F'")['totvotpoll'])
```

```python
tot_votes_fem = fdf.groupby('year')[['totvotpoll']].sum()
tot_votes_fem = tot_votes_fem.reset_index()
tot_votes_fem
```

```python
tot_voters = df.groupby('year')[['totvotpoll']].sum()
tot_voters = tot_voters.reset_index()
tot_voters
```

```python
overall_yoy = pd.DataFrame(columns=['year', 'growth'])
for i, rw in tot_voters.iterrows():
#     print(rw['year'])
#     print(i)
    if i == 0:
#         print('------------------')
        continue
    else:
#         print(rw['year'])
        a = tot_votes_fem.loc[tot_votes_fem['year'] == rw['year']]['totvotpoll'].iloc[0]
        b = tot_voters.loc[tot_voters['year'] == rw['year']]['totvotpoll'].iloc[0]
#         print(a)
#         print(b)
#         print(a/b)
#         print(tot_voters.loc[i - 1]['year'])
        a1 = tot_votes_fem.loc[tot_votes_fem['year'] == tot_voters.loc[i - 1]['year']]['totvotpoll'].iloc[0]
        b1 = tot_voters.loc[tot_voters['year'] == tot_voters.loc[i - 1]['year']]['totvotpoll'].iloc[0]
#         print(a1)
#         print(b1)
#         print(a1/b1)
#         print(((a/b - a1/b1)/(a1/b1)) * 100)
        overall_yoy = overall_yoy.append({'year': rw['year'], 'growth': ((a/b - a1/b1)/(a1/b1)) * 100}, ignore_index=True)
#         print(rw['year'])
#         print(df.loc[i - 1]['year'])
#     print("-----------------")
overall_yoy
```

```python
overall_yoy.plot(x='year', y='growth', kind='bar') #df.plot(x='col_name_1', y='col_name_2')
```

```python
right_df = fdf.groupby(['year', 'st_name'])[['totvotpoll']].sum()
right_df
```

```python
left_df = df.groupby(['year', 'st_name'])[['totvotpoll']].sum()
left_df
```

```python
state_df = pd.merge(left_df, right_df, how='left', on=['year', 'st_name'])
state_df = state_df.fillna(0)
state_df
```

```python
prev_yr = {}
for i, rw in tot_voters.iterrows():
    if i == 0:
        continue
    else:
        prev_yr[rw['year']] = tot_voters.loc[i - 1]['year']
prev_yr
```

```python
import numpy
state_yoy = pd.DataFrame(columns=['year', 'state', 'growth'])
# state_df.loc[(prev_yr[1980], 'Andhra Pradesh')]
for i, rw in state_df.iterrows():
    if rw.name[0] == 1977:
#     if rw.name[0] != 1984 or i[1] != 'Andaman & Nicobar Islands':
        continue
    elif (prev_yr[i[0]], i[1]) in state_df.index:
#         print(i[0])
#         print(state_df.loc[(prev_yr[i[0]], i[1])])
        a = rw['totvotpoll_y']
        b = rw['totvotpoll_x']
#         print(a)
#         print(b)
        a1 = state_df.loc[(prev_yr[i[0]], i[1])]['totvotpoll_y']
        b1 = state_df.loc[(prev_yr[i[0]], i[1])]['totvotpoll_x']
#         print(a1)
#         print(b1)
        if a1 == 0:
            state_yoy = state_yoy.append({'year': i[0], 'state': i[1], 'growth': a / b * 100}, ignore_index=True)
        else:
            state_yoy = state_yoy.append({'year': i[0], 'state': i[1], 'growth': ((a/b - a1/b1)/(a1/b1)) * 100}, ignore_index=True)
state_yoy
```

```python
df2009 = state_yoy.query("year == 2009")
df2014 = state_yoy.query("year == 2014") # state_yoy.loc[state_yoy['year'] == '2014']
df2014.plot(x='state', y='growth', kind='bar')
```

```python
left_df1 = df.groupby(['year', 'region'])[['totvotpoll']].sum()
right_df1 = fdf.groupby(['year', 'region'])[['totvotpoll']].sum()
region_df = pd.merge(left_df1, right_df1, how='left', on=['year', 'region'])
region_df = region_df.fillna(0)
region_df
```

```python
import numpy
region_yoy = pd.DataFrame(columns=['year', 'region', 'growth'])
for i, rw in region_df.iterrows():
    if rw.name[0] == 1977:
        continue
    else:
#         print(rw.name)
        a = rw['totvotpoll_y']
        b = rw['totvotpoll_x']
        a1 = region_df.loc[(prev_yr[i[0]], i[1])]['totvotpoll_y']
        b1 = region_df.loc[(prev_yr[i[0]], i[1])]['totvotpoll_x']
        region_yoy = region_yoy.append({'year': i[0], 'region': i[1], 'growth': ((a/b - a1/b1)/(a1/b1)) * 100}, ignore_index=True)
region_yoy.set_index(['year', 'region'])
```

```python
# region_yoy.plot(x='year', y='growth, region')
region_piv = region_yoy.pivot(index='year', columns='region', values='growth')

region_piv.plot(kind='bar')
```
